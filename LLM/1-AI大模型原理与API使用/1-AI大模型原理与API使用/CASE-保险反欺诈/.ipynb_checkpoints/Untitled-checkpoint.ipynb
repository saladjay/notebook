```python
import pandas as pd

# 加载训练集和测试集数据
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

# 检查训练集数据
print("训练集数据:")
print(train_data.head())

# 检查测试集数据
print("测试集数据:")
print(test_data.head())

# 提取特征和目标变量
X_train = train_data.drop('fraud_reported', axis=1)  # 特征变量
y_train = train_data['fraud_reported']  # 目标变量

# 检查提取后的特征和目标变量
print("训练集特征变量:")
print(X_train.head())
print("训练集目标变量:")
print(y_train.head())

# 保存测试集的特征变量，不包括目标变量
X_test = test_data  # 测试集特征变量

# 检查保存的测试集特征变量
print("测试集特征变量:")
print(X_test.head())
```

```python
# 合并训练集和测试集
data = pd.concat([train_data, test_data], ignore_index=True)
data
```

```python
# 检查数据的缺失值
missing_values = data.isnull().sum()
print("缺失值统计:")
print(missing_values)

# 查找字段的唯一值个数
unique_values = data.nunique()
print("字段的唯一值个数:")
print(unique_values)
```

```python
# 去除 _c39 字段
data = data.drop('_c39', axis=1)
data
```

```python
# 将 policy_bind_date 列转换为日期类型
data['policy_bind_date'] = pd.to_datetime(data['policy_bind_date'])

# 计算 policy_bind_date_diff
min_bind_date = data['policy_bind_date'].min()
data['policy_bind_date_diff'] = (data['policy_bind_date'] - min_bind_date).dt.days
```

```python
# 将 incident_date 和 policy_bind_date 列转换为日期类型
data['incident_date'] = pd.to_datetime(data['incident_date'])
data['policy_bind_date'] = pd.to_datetime(data['policy_bind_date'])

# 计算 incident_date_diff
min_bind_date = data['policy_bind_date'].min()
data['incident_date_diff'] = (data['incident_date'] - min_bind_date).dt.days
```

```python
# 去除 policy_bind_date 和 incident_date 字段
data = data.drop(['policy_bind_date', 'incident_date'], axis=1)
data
```

```python
# 选择字符串类型的字段
string_columns = data.select_dtypes(include='object').columns
data[string_columns]
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# 标签编码转换
label_encoder = LabelEncoder()
for column in string_columns:
    data[column + '_encoded'] = label_encoder.fit_transform(data[column])
data
```

```python
# 去除字符串类型的字段
data = data.drop(string_columns, axis=1)
data
```

```python
# 切分回 train_data 和 test_data
train_data = data.iloc[:len(train_data), :]
test_data = data.iloc[len(train_data):, :]
train_data
```

```python
test_data
```

```python
import lightgbm as lgb

# 提取特征和目标变量
X_train = train_data.drop('fraud_reported', axis=1)  # 训练集特征变量
y_train = train_data['fraud_reported']  # 训练集目标变量

# 去除测试集中的目标变量字段
X_test = test_data.drop('fraud_reported', axis=1)  # 测试集特征变量

# 创建 LightGBM 数据集
train_dataset = lgb.Dataset(X_train, label=y_train)

# 定义 LightGBM 参数
params = {
    'objective': 'binary',
    'metric': 'binary_logloss'
}

# 训练 LightGBM 模型
model = lgb.train(params, train_dataset)

# 对测试集进行预测
y_pred = model.predict(X_test)

# 打印预测结果
print("预测结果:")
print(y_pred)
```

```python
# 读取 sampleSubmission.csv 文件
submission_data = pd.read_csv('sampleSubmission.csv')

# 将预测结果赋值给 fraud_reported 字段
submission_data['fraud_reported'] = y_pred

# 保存结果为 result.csv 文件
submission_data.to_csv('result.csv', index=False)
```
