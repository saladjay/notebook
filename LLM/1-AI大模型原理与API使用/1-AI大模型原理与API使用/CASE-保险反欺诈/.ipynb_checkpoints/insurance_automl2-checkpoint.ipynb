```python
import pickle

with open('./data.pkl', 'rb') as file:
    data = pickle.load(file)
data
```

```python
# 训练集、测试集切分
train = data[data['fraud_reported'].notnull()]
test = data[data['fraud_reported'].isnull()]
test
```

```python
from autogluon.tabular import TabularPredictor

# 创建AutoML模型
model = TabularPredictor(label="fraud_reported", path="model")

# 可以指定训练时长
#model.fit(train, time_limit=300)
#model = TabularPredictor(label, eval_metric=metric, path=save_path).
model.fit(train, presets='best_quality')

# 查看Leaderboard
model.leaderboard()
```

```python
# AutoML预测
y_pred = model.predict_proba(test)
y_pred = y_pred.loc[:,1]
y_pred
```

```python
import pandas as pd

result = pd.read_csv('./sampleSubmission.csv')
result['fraud_reported'] = y_pred.values
result.to_csv("./baseline_automl1.csv", index=False)
result
```
