## 保险反欺诈
* 特征重要性排序（细粒度）

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import lightgbm as lgb
import pickle

with open('./data.pkl', 'rb') as file:
    data = pickle.load(file)

data
```

```python
# 处理日期类型字段：policy_bind_date, incident_date
#data.select_dtypes(include="O")
# 日期类型格式转换
data['policy_bind_date'] = pd.to_datetime(data['policy_bind_date'])
data['incident_date'] = pd.to_datetime(data['incident_date'])

# 观察最大值，最小值
data['policy_bind_date'].min() # 1990-01-08
data['policy_bind_date'].max() # 2015-02-22

data['incident_date'].min() # 2015-01-01
data['incident_date'].max() # 2015-03-01

date_base = data['policy_bind_date'].min()
# 转换为date_diff
data['policy_bind_date_diff'] = (data['policy_bind_date'] - date_base).dt.days
data['incident_date_diff'] = (data['incident_date'] - date_base).dt.days
# 构造新特征：incident_policy_bind_diff
data['incident_policy_bind_diff'] = (data['incident_date'] - data['policy_bind_date']).dt.days
data[['policy_bind_date_diff', 'incident_date_diff', 'incident_policy_bind_diff']]
```

```python
# 去掉数字 和 空格
def remove_digit(string):
    newstring = ''.join([x for x in string if (not x.isdigit() and x!=' ')])
    return newstring

data['incident_location'] = data['incident_location'].apply(remove_digit)
data['incident_location'].value_counts()
```

```python
# 去掉不用的字段
data.drop(['policy_bind_date','incident_date','policy_number'],axis=1,inplace=True)
```

```python
data.columns
data['incident_severity']
```

```python
# 将分类特征全部展开
all_data = pd.get_dummies(data)
```

```python
# 数据集切分
train = all_data[all_data['fraud_reported'].notnull()]
test = all_data[all_data['fraud_reported'].isnull()]
```

```python
import xgboost as xgb

model_xgb = xgb.XGBClassifier(
            max_depth=6, learning_rate=0.001, n_estimators=2000, 
            objective='binary:logistic', #tree_method='gpu_hist', 
            subsample=0.8, colsample_bytree=0.8, 
            min_child_samples=3, eval_metric='auc', reg_lambda=0.5
        )
model_xgb.fit(train.drop(['fraud_reported'], axis=1), train['fraud_reported'],
    eval_set=[(train.drop(['fraud_reported'], axis=1), train['fraud_reported'])], early_stopping_rounds=50)
```

```python
# 查看特征重要性
from xgboost import plot_importance

fig, ax = plt.subplots(figsize=(10,8))
plot_importance(model_xgb, max_num_features=20, ax=ax)
plt.show()
```
