```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
import lightgbm as lgb

# 数据加载
train = pd.read_csv("./train.csv")
test = pd.read_csv("./test.csv")
sample = pd.read_csv('./sampleSubmission.csv')
train
```

```python
data = pd.concat([train, test], axis=0)
data.index = range(len(data))
data
```

```python
for col in data.columns:
    print(col, data[col].nunique())
```

```python
# 这里发现 _c39字段为空
data['_c39'].value_counts()
```

```python
# 去掉空值的字段
data.drop('_c39',axis=1,inplace=True)
```

```python
# 发现没有缺失值
data.isnull().sum()
```

```python
data.to_pickle('./data.pkl')
```

```python
cat_columns = data.select_dtypes(include='O').columns

# 标签编码
le = LabelEncoder()
le_count = 0

for col in cat_columns:
    if data[col].nunique() <= 2:
        data[col] = le.fit_transform(data[col])
        le_count += 1
            
print('{}列 进行了标签编码'.format(le_count))
```

```python
# 查看相关性系数
plt.figure(figsize=(16, 16))
sns.heatmap(data.corr(), annot=True, linewidths=.5)
```

```python
# 查看唯一值个数
colum_name =[]
unique_value=[]

for col in data.select_dtypes(include='O').columns:
    colum_name.append(str(col)) 
    unique_value.append(data[col].nunique())

table= pd.DataFrame()
table['Col_name'] = colum_name
table['Value']= unique_value
            
table=table.sort_values('Value',ascending=False)
table
```

```python
# 包含空缺的字段：property_damage, police_report_available
data['property_damage'] = data['property_damage'].map({'YES': 1, "NO": 0, '?': 2})
data['police_report_available'] = data['police_report_available'].map({'YES': 1, "NO": 0, '?': 2})
```

```python
# 处理日期类型字段：policy_bind_date, incident_date
#data.select_dtypes(include="O")
# 日期类型格式转换
data['policy_bind_date'] = pd.to_datetime(data['policy_bind_date'])
data['incident_date'] = pd.to_datetime(data['incident_date'])

# 观察最大值，最小值
data['policy_bind_date'].min() # 1990-01-08
data['policy_bind_date'].max() # 2015-02-22

data['incident_date'].min() # 2015-01-01
data['incident_date'].max() # 2015-03-01

date_base = data['policy_bind_date'].min()
# 转换为date_diff
data['policy_bind_date_diff'] = (data['policy_bind_date'] - date_base).dt.days
data['incident_date_diff'] = (data['incident_date'] - date_base).dt.days
# 构造新特征：incident_policy_bind_diff
data['incident_policy_bind_diff'] = (data['incident_date'] - data['policy_bind_date']).dt.days

data[['policy_bind_date_diff', 'incident_date_diff', 'incident_policy_bind_diff']]
```

```python
data[['incident_location','auto_model','insured_occupation']]
```

```python
# 去掉incident_location中的数字
data['incident_location'].value_counts()
#data['insured_occupation'].value_counts()
#data.info()
```

```python
# 去掉数字 和 空格
def remove_digit(string):
    newstring = ''.join([x for x in string if (not x.isdigit() and x!=' ')])
    return newstring

data['incident_location'] = data['incident_location'].apply(remove_digit)
data['incident_location'].value_counts()
```

```python
# 去掉不用的字段
data.drop(['policy_bind_date','incident_date','policy_number'],axis=1,inplace=True)
```

```python
# 标签编码
from sklearn.preprocessing import LabelEncoder
cat_cols = data.select_dtypes(include='O').columns
for col in cat_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
data
```

```python
# 数据集切分
train = data[data['fraud_reported'].notnull()]
test = data[data['fraud_reported'].isnull()]
```

```python
import lightgbm as lgb

model_lgb = lgb.LGBMClassifier(
            num_leaves=2**5-1, reg_alpha=0.25, reg_lambda=0.25, objective='binary',
            max_depth=-1, learning_rate=0.005, min_child_samples=3, random_state=2022,
            n_estimators=2000, subsample=1, colsample_bytree=1,
        )
model_lgb.fit(train.drop(['fraud_reported'], axis=1), train['fraud_reported'])
y_pred_lgb = model_lgb.predict_proba(test.drop(['fraud_reported'], axis=1))[:, 1]
#y_pred_lgb
```

```python
from sklearn.metrics import roc_auc_score
help(roc_auc_score)
```

```python
# 五折交叉验证，模型融合
from sklearn.model_selection import StratifiedKFold

model_lgb = lgb.LGBMClassifier(
        num_leaves=2**5-1, reg_alpha=0.25, reg_lambda=0.25, objective='binary',
        max_depth=-1, learning_rate=0.001, min_child_samples=3, random_state=2022,
        n_estimators=10000, subsample=1, colsample_bytree=1, metric='auc'
    )

# 五折交叉验证
sk = StratifiedKFold(n_splits=5, shuffle=True, random_state=2022)
index = 0
y_pred_list = [0 for i in range(300)]
for train_idx, valid_idx in sk.split(train, train['fraud_reported']):
    index = index + 1
    # 子模型数据
    train_data = train.iloc[train_idx]
    valid_data = train.iloc[valid_idx]
    #print(train_data)
    # 子模型训练
    model_lgb.fit(train_data.drop(['fraud_reported'], axis=1), train_data['fraud_reported'],
        eval_set=[(train_data.drop(['fraud_reported'], axis=1), train_data['fraud_reported']),
                  (valid_data.drop(['fraud_reported'], axis=1), valid_data['fraud_reported'])], 
        eval_names = ['train','valid'], eval_metric='auc', early_stopping_rounds=500)
    # 模型评估
    y_pred_lgb = model_lgb.predict_proba(valid_data.drop(['fraud_reported'], axis=1))[:, 1]
    print('子模型{} AUC:{}'.format(index, roc_auc_score(valid_data['fraud_reported'], y_pred_lgb)))
    
    # 模型预测
    y_pred_lgb = model_lgb.predict_proba(test.drop(['fraud_reported'], axis=1),
                                        num_iteration=model_lgb.best_iteration_)[:, 1]
    #y_pred_list.append(y_pred_lgb)
    y_pred_list += y_pred_lgb / 5
```

```python
result = pd.read_csv('./sampleSubmission.csv')
result['fraud_reported'] = y_pred_list
result.to_csv('./baseline_lgb_skf51.csv', index=False)
result
```

```python
# 五折交叉验证，模型融合
from sklearn.model_selection import StratifiedKFold
import catboost as cb

model_cb = cb.CatBoostClassifier(iterations=10000, 
                              depth=7, 
                              learning_rate=0.001, 
                              loss_function='Logloss',
                              eval_metric='AUC',
                              logging_level='Verbose', 
                              metric_period=200)

# 五折交叉验证
sk = StratifiedKFold(n_splits=5, shuffle=True, random_state=2022)
index = 0
y_pred_list = [0 for i in range(300)]
for train_idx, valid_idx in sk.split(train, train['fraud_reported']):
    index = index + 1
    # 子模型数据
    train_data = train.iloc[train_idx]
    valid_data = train.iloc[valid_idx]
    # 子模型训练
    model_cb.fit(train_data.drop(['fraud_reported'], axis=1), train_data['fraud_reported'],
        eval_set=[(train_data.drop(['fraud_reported'], axis=1), train_data['fraud_reported']),
                  (valid_data.drop(['fraud_reported'], axis=1), valid_data['fraud_reported'])], 
                early_stopping_rounds=50)
    # 模型评估
    y_pred_cb = model_cb.predict_proba(valid_data.drop(['fraud_reported'], axis=1))[:, 1]
    print('子模型{} AUC:{}'.format(index, roc_auc_score(valid_data['fraud_reported'], y_pred_cb)))
    
    # 模型预测
    y_pred_cb = model_cb.predict_proba(test.drop(['fraud_reported'], axis=1))[:, 1]
    y_pred_list += y_pred_cb / 5
```
