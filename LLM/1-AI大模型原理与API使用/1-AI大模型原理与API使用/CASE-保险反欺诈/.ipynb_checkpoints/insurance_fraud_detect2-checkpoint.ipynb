```python
import pandas as pd

# 数据加载
train = pd.read_csv('./train.csv')
train
```

```python
test = pd.read_csv('./test.csv')
test
```

```python
data = pd.concat([train, test], axis=0)
data.index = range(len(data))
data
```

```python
# 查看特征的唯一值个数
for col in data.columns:
    print(col, data[col].nunique())
```

```python
# 发现 _c39字段为空，去掉空值的字段
#data['_c39'].value_counts()
data[['_c39']]
data.drop('_c39', axis=1, inplace=True)
data
```

```python
# 缺失值个数
data.isnull().sum()
```

```python
# 日期类型： policy_bind_date, incident_date
# 找到分类特征
cat_columns = data.select_dtypes(include='O').columns
cat_columns
```

```python
# 找到分类特征里面，是否有唯一个数为2
for col in cat_columns:
    if data[col].nunique() <=2:
        print(col, data[col].nunique())
```

```python
data[['insured_sex']]
```

```python
from sklearn.preprocessing import LabelEncoder
# 标签编码
le = LabelEncoder()
data['insured_sex'] = le.fit_transform(data['insured_sex'])
data[['insured_sex']]
```

```python
# 查看特征的唯一值个数
columns_name = []
unique_value = []

for col in data.select_dtypes(include='O').columns:
    columns_name.append(col)
    unique_value.append(data[col].nunique())
table = pd.DataFrame()
table['col'] = columns_name
table['value'] = unique_value
table
```

```python
table.sort_values(by=['value'], ascending=False)
```

```python
#data[['incident_date']]
#data.select_dtypes(include='O')

data['property_damage'].value_counts()
data['property_damage'] = data['property_damage'].map({'YES': 1, 'NO':0, '?': 2})
data['property_damage'].value_counts()
```

```python
data['police_report_available'].value_counts()
data['police_report_available'] = data['police_report_available'].map({'YES': 1, 'NO':0, '?': 2})
data['police_report_available'].value_counts()
```

```python
# include='O' Object
# 日期类型： policy_bind_date, incident_date
#data[['policy_bind_date', 'incident_date']]
#data.info()
#data.select_dtypes(include='O')

# 日期类型转换
data['policy_bind_date'].min() # 1990-01-08
data['incident_date'].min() # 2015-01-01
# 转换为日期格式
data['policy_bind_date'] = pd.to_datetime(data['policy_bind_date'])
data['incident_date'] = pd.to_datetime(data['incident_date'])

# 计算日期diff
data['policy_bind_date_diff'] = (data['policy_bind_date'] - pd.to_datetime('1990-01-08')).dt.days
data['incident_date_diff'] = (data['incident_date'] - pd.to_datetime('2015-01-01')).dt.days

data[['policy_bind_date_diff', 'incident_date_diff']]
```

```python
data.drop(['policy_number', 'policy_bind_date', 'incident_date'], axis=1, inplace=True)
data
```

```python
# 去掉日期列
data.drop(['policy_bind_date', 'incident_date'], axis=1, inplace=True)
data
```

```python
# 训练集、测试集切分
train = data[data['fraud_reported'].notnull()]
train
```

```python
test = data[data['fraud_reported'].isnull()]
test
```

```python
# 模型训练
import lightgbm as lgb
model_lgb = lgb.LGBMClassifier(
            num_leaves=2**5-1, reg_alpha=0.25, reg_lambda=0.25, objective='binary',
            max_depth=-1, learning_rate=0.005, min_child_samples=3, random_state=2022,
            n_estimators=2000, subsample=1, colsample_bytree=1)
model_lgb.fit(train.drop(['fraud_reported'], axis=1), train['fraud_reported'])
```

```python
#y_pred = model_lgb.predict(test.drop(['fraud_reported'], axis=1))
y_pred = model_lgb.predict_proba(test.drop(['fraud_reported'], axis=1))[:, 1]
y_pred
```

```python
result = pd.read_csv('./sampleSubmission.csv')
result['fraud_reported'] = y_pred
result.to_csv('baseline_lgb2.csv', index=False)
result
```
