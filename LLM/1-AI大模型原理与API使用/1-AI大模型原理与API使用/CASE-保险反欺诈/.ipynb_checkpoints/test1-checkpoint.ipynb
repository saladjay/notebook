```python
import pandas as pd

# 加载训练集数据
train_data = pd.read_csv('train.csv')

# 加载测试集数据
test_data = pd.read_csv('test.csv')

# 查看训练集数据的前几行
print("训练集数据示例:")
print(train_data.head())

# 查看测试集数据的前几行
print("测试集数据示例:")
print(test_data.head())
```

```python
# 将训练集和测试集数据合并
data = pd.concat([train_data, test_data], ignore_index=True)

# 查看合并后的数据集
print("合并后的数据集:")
print(data.head())
```

```python
# 查找缺失值
missing_values = data.isnull().sum()
print("缺失值统计:")
print(missing_values)

# 查找字段的唯一值个数
unique_values = data.nunique()
print("字段的唯一值个数:")
print(unique_values)
```

```python
# 去掉字段 '_c39'
data = data.drop('_c39', axis=1)

# 查看更新后的数据集
print("更新后的数据集:")
print(data.head())
```

```python
# 筛选字符串字段
string_columns = data.select_dtypes(include=['object']).columns.tolist()

# 赋值给 cat_columns
cat_columns = string_columns
# 打印字符串字段
print("字符串字段:")
print(cat_columns)
```

```python
# 计算每个字符串字段的唯一值个数
unique_counts = data[string_columns].nunique().sort_values(ascending=False)

# 按照唯一值个数从大到小排序
unique_counts_sorted = unique_counts.sort_values(ascending=False)

# 打印唯一值个数和排序结果
print("字符串字段的唯一值个数和排序结果:")
print(unique_counts_sorted)
```

```python
data['property_damage'].value_counts()
```

```python
# 将 'NO' 替换为 0，'YES' 替换为 1，'?' 替换为 2
data['property_damage'] = data['property_damage'].replace({'NO': 0, 'YES': 1, '?': 2})
data['property_damage'].value_counts()
```

```python
# 将 'policy_bind_date' 和 'incident_date' 转换为日期类型
data['policy_bind_date'] = pd.to_datetime(data['policy_bind_date'])
data['incident_date'] = pd.to_datetime(data['incident_date'])
```

```python
# 计算 'policy_bind_date' 中的最小日期
min_policy_bind_date = data['policy_bind_date'].min()

# 创建 'policy_bind_date_diff' 字段
data['policy_bind_date_diff'] = (data['policy_bind_date'] - min_policy_bind_date).dt.days
```

```python
# 创建 'incident_date_diff' 字段
data['incident_date_diff'] = (data['incident_date'] - min_policy_bind_date).dt.days
```

```python
# 创建 'incident_date_policy_bind_date_diff' 字段
data['incident_date_policy_bind_date_diff'] = data['incident_date_diff'] - data['policy_bind_date_diff']
```

```python
# 删除原始的日期字段 'policy_bind_date' 和 'incident_date'
data = data.drop(['policy_bind_date', 'incident_date'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

# 筛选字符串字段
string_columns = data.select_dtypes(include=['object']).columns.tolist()

# 对字符串字段进行标签编码
label_encoder = LabelEncoder()
for column in string_columns:
    data[column] = label_encoder.fit_transform(data[column])
data
```

```python
# 根据索引将数据集拆分回训练集和测试集
train_data = data.loc[:train_data.shape[0]-1, :]
test_data = data.loc[train_data.shape[0]:, :]
train_data
```

```python
import lightgbm as lgb
from sklearn.metrics import roc_auc_score

# 定义特征列和目标列
features = train_data.drop('fraud_reported', axis=1)
target = train_data['fraud_reported']

# 构建LightGBM数据集
train_dataset = lgb.Dataset(features, label=target)

# 定义LightGBM参数
params = {
    'objective': 'binary',
    'metric': 'binary_logloss'
    # 添加其他参数...
}

# 训练LightGBM模型
model = lgb.train(params, train_dataset, num_boost_round=100)

# 在测试集上进行预测
#test_features = test_data  # 测试集的特征列
########################################
# 去掉目标列 'fraud_reported'
test_features = test_data.drop('fraud_reported', axis=1)
y_pred_proba = model.predict(test_features)

# 打印预测概率值
print("测试集预测概率值:")
print(y_pred_proba)

# 可选：将概率值转换为二分类标签
y_pred = [1 if pred >= 0.5 else 0 for pred in y_pred_proba]

"""
# 可选：计算AUC作为模型性能评估指标
target_test = test_data['fraud_reported']  # 测试集的目标列
auc = roc_auc_score(target_test, y_pred_proba)
print("测试集AUC:", auc)
"""
```

```python
# 读取sampleSubmission.csv文件
submission_data = pd.read_csv('sampleSubmission.csv')

# 将 'fraud_reported' 字段设置为预测概率值 y_pred_proba
submission_data['fraud_reported'] = y_pred_proba

# 保存结果为result_baseline.csv文件
submission_data.to_csv('result_baseline.csv', index=False)
```

```python
import matplotlib.pyplot as plt
# model是我们的模型，我想从模型中洞察重要的特征，将这些特征重要性进行可视化

# 获取特征重要性
feature_importance = model.feature_importance()
feature_names = model.feature_name()

# 创建特征重要性的数据框
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importance})

# 按照重要性降序排序
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# 可视化特征重要性
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance')
plt.show()
```
