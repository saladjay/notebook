```python
import torch
import numpy as np
from torch import nn
a1 = torch.randn(1, 64, 4)
a2 = torch.randn(1, 64, 2)
a3 = torch.randn(1, 64, 1)
conv = nn.Conv2d(16, 1, 1, bias=False).requires_grad_(False)
conv.eval()
```
